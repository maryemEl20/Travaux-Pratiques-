```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#charger les donnees a partir d'un fichier csv
data = pd.read_csv('Salary_Data.csv')
```

```python
# les 5 premiéres instances
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#preparing the data
#x comme variables d'entrées
#y comme cible
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
# Créer un objet de régression linéaire
regression = LinearRegression()
```

```python
# Entraîner les donnees
regression.fit(X, y)
```

```python
#Mettre les points sur un graph
# Tracer la ligne de régression sur les données d'entraînement
plt.scatter(X, y , label='Données d\'entraînement')
plt.xlabel('Years Experience')
plt.ylabel('Salary)')
plt.title('Régression linéaire sur les données')
plt.legend()
plt.show()
```

```python
#retrouver les coefficines et l'ordonnées a l origine
#Ces résultats vous donnent les paramètres de votre modèle de régression linéaire

# Affichage des coefficients du modèle
print("Coefficients du modèle:", regression.coef_)
print("Intercept du modèle:", regression.intercept_)
          

```

```python
# Faire de nouvelles prédictions avec le modèle entraîné
predictions = regression.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, color='red')
```

```python
# Calculer la précision du modèle
print("Précision du modèle :", regression.score(X, y))
```
