## *Question 2*

```python
#
import numpy as np
import matplotlib.pyplot as plt

def compute_error_for_line_given_point(b,w,points):
    totalerror=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        totalerror+=(y-(w*x+b))**2
    return totalerror/(2*float(len(points)))
```

## *Question 3*

```python
#
def step_gradent(b_current,w_current,points,learnrate):
    b_gradient=0
    w_gradient=0
    N=float(len(points))

    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        #
        b_gradient+=(1/N)*((w_current*x+b_current)-y)#
        w_gradient+=(1/N)*x*((w_current*x+b_current)-y)
    #renew
    new_b=b_current-(learnrate*b_gradient)
    new_w=w_current-(learnrate*w_gradient)
    return [new_b,new_w]
```

## *Question 4*

```python
def gradient_descent_runner(points,starting_b,starting_w,learning_rate,num_iteration):
    b=starting_b
    w=starting_w
    cost_history = np.zeros(num_iteration)
    for i in range(num_iteration):
        b,w=step_gradent(b,w,np.array(points),learning_rate)
        print("Iteration Numero :",i)
        print("Fonction du cout = ", compute_error_for_line_given_point(b,w,np.array(points)))
        cost_history[i] = compute_error_for_line_given_point(b,w,np.array(points))
        print("w=", w)
        print("b=", b)
    plt.plot(range(num_iteration), cost_history)
    plt.show()
    return [b,w]
```

## *Question 5*

```python
# Hyperparameter : learning_rate = 0.0001 et nombre_itération = 100 :
def run():

    # Initialize one hundred points
    x=np.linspace(1,10,100)
    y=np.linspace(1,10,100)

    array=[]

    for i in np.arange(100):
        for j in np.arange(100):
            p=[x[i],y[i]]
            array.append(p)


    learning_rate=0.0001
    init_b=0
    init_w=0
    num_iteration=100

    [b,w]=gradient_descent_runner(array,init_b,init_w,learning_rate,num_iteration)

    plt.figure()

    plt.scatter(x, y)
    y2=b+w*x

    plt.scatter(x,y2)
    print("last_b=", b)
    print("last_w=", w)

if __name__=="__main__":
    run()
```

## *Question 1*

```python
# Importer les librairies
import numpy as np
import matplotlib.pyplot as plt
from   sklearn import datasets
```

## *Question 2*

```python
# Créer votre dataset de régression
x, y = datasets.make_regression(n_samples=100, n_features=1, noise=10)
```

## *Question 3*

```python
# Affichage du dataset
plt.scatter(x, y)
plt.xlabel('(X)')
plt.ylabel('(y)')
plt.show()
```

## *Question 4*

```python
from sklearn.linear_model import SGDRegressor
# Définir le modèle avec SGDRegressor
model = SGDRegressor(max_iter=100, learning_rate='constant', eta0=0.0001, random_state=42)
```

## *Question 5*

```python
# Entraîner le modèle
model.fit(x, y)
```

## *Question 6*

```python
# Calculer la précision du modèle
precision = model.score(x, y)

print("Précision du modèle  :", precision)
```

## *Question 7*

```python
# Faire de nouvelles prédictions
predictions = model.predict(x)
# Tracer les résultats
plt.scatter(x, y, label='Données réelles')
plt.plot(x, predictions, color='red', label='Prédictions')
plt.xlabel('(X)')
plt.ylabel('(y)')
plt.legend()
plt.show()
```

## *Question 8*

```python
#Refaire le même travail en entraînant votre modèle sur 1000 itérations avec un Learning rate de 0.001.
model_1000 = SGDRegressor(max_iter=1000, learning_rate='constant', eta0=0.001, random_state=42)
model_1000.fit(x,y)
# Calculer la précision du modèle
precision = model_1000.score(x, y)
print("Précision du modèle  :", precision)

# Tracer les résultats
plt.scatter(x, y, label='Données réelles')
plt.plot(x,  model_1000.predict(x), color='red', label='Prédictions')
plt.xlabel('(X)')
plt.ylabel('(y)')
plt.legend()
plt.show()
```
